# 模型特点
1. 双路结构：
◦ Wide部分：处理高维交叉特征，增强记忆能力
◦ Deep部分：通过嵌入层学习特征表示，提升泛化能力
2. 联合训练：端到端同时优化wide和deep部分
3. 特征工程：需预先设计交叉特征（代码中通过set_crossed_features生成）
