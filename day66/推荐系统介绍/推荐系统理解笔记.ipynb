# 推荐系统理解笔记
## 一、推荐系统简介
个性化推荐（推荐系统）是 AI 成功落地的重要分支，已成为互联网产品标配，在电商（淘宝 / 京东）、资讯（今日头条 / 微博）、音乐（网易云音乐 / QQ 音乐）、短视频（抖音 / 快手）、大数据风控、广告推荐等领域均为核心组件。
## 二、推荐系统产生背景
源于信息过载与用户需求不明确的问题，信息获取方式经历了以下发展阶段：
分类目录（1990s）：如 Hao123、Yahoo、3721，仅覆盖少量热门网站。
搜索引擎（2000s）：如 Google、Baidu，用户通过搜索词明确需求。
推荐系统（2010s）：无需用户提供明确需求，通过分析用户历史行为建模其兴趣，主动推荐满足需求的信息。
## 三、什么是推荐系统
当没有明确需求的用户访问服务，且服务的物品对用户构成信息过载时，系统通过一定规则对物品排序，并将排在前面的物品展示给用户，这样的系统就是推荐系统。
## 四、推荐系统与搜索引擎的对比
对比项	搜索	推荐
行为方式	主动	被动
意图	明确	模糊
个性化	弱	强
流量分布	马太效应	长尾效应
目标	快速满足	持续服务
评估指标	简明	复杂
## 五、推荐系统的作用
高效连接用户和物品，发现长尾商品。
留住用户和内容生产者（销售商品），实现商业目标。
## 六、推荐系统的工作原理
社会化推荐：向朋友咨询，让好友推荐物品。
基于内容的推荐：类似打开搜索引擎，输入喜欢的演员名字，查看返回结果中未看过的电影。
基于流行度的推荐：如查看票房排行榜。
基于协同过滤的推荐：找到与自己历史兴趣相似的用户，查看他们最近看的电影。
## 七、推荐系统的应用场景
feed 流、信息流等。
## 八、推荐系统和 Web 开发项目的区别
本质目标不同：
Web 项目（开发类型项目）：处理复杂逻辑、高并发，实现高可用，为用户提供稳定的信息流通服务。
推荐系统：追求指标增长，如留存率、阅读时间、GMV（电商网站成交金额）、视频网站 VV（Video View）。
思维方式不同：
Web 项目：对结果有确定预期。
推荐系统：结果是概率问题（可能一开始表现差，后来表现好）。
